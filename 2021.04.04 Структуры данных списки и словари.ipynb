## <h2><center>Список</center></h2>
Как только мы начинаем выходить за рамки написания простейших программ у нас возниает потребность в хранении коллекций или наборов данных. Для этого в языке Python предусмотрены 2 очень мощных по своим возможностям типа данных: списки и словари.




**Список** в Python — это упорядоченная изменяемая коллекция объектов произвольных типов.




В этом определении важна каждое слово. Мы будем к нему возвращаться по мере иллюстрации возможностей и свойств списков.



* Когда мы называем список коллекцией, мы имеем в виду, что в одном списке может храниться множество объектов (например, чисел или строковых величин) 


* Говоря о том, что списки представляют собой упорядоченные коллекции, мы обращаем внимание на то, что каждый элемент в списке имеет свой порядковый номер.


* Упоминание об изменяемости списков означает, что в процессе исполнения кода можно изменять состав элементов списка и последовательность их расположения.


* Выражение «произвольных типов» означает, что в списке могут храниться данные, относящиеся к любым типам из тех, с которыми работает Python. *То есть это могут быть также другие списки и словари*

### Списки и циклы

Списки хорошо сочетаются с циклами. Такое сочетание возможно благодаря оператору in, используемому в цикле for

```python
#У нас есть фургон, в котором мы перевозим грузы. Установлено ограничение на максимальный вес груза в 100кг
items = [1, 5, 76, 12, 123, 333, 5, 6, 7]
ok_items = []

for gruz in items:
    print('Груз весит ' + str(gruz) +  ' килограмм')
    if gruz<100:
        ok_items.append(gruz) 
        print('Груз прошел проверку')
    else:
        print('Груз слишком большой')
ok_items
```

**Рубрика "реальные задачи"**. Вспомним наш пример с шахматной доской. Давайте смоделируем случайную расстановку фигур на шахматной доске

```python
import random
```

```python
#проинициализируем шахматную доску
chess_board = []#задаем пустой список
for i in range(8):#двигаемся сначала по строкам нашей таблицы
    row = []#создаем в каждой строке по пустому столбцу
    for j in range(8):#столбец заполняем пробелами
        row.append(' ')
    chess_board.append(row)#добавляем столбец 
chess_board
```

```python
#расставим случайным образом несколько фигур
chess_pieces = ['king ', 'bishop', 'bishop', 'queen', 'knight']#создадим метки фигур, которые мы хотим разместить на нашей доске
for piece in chess_pieces: #перебираем все фигуры
    i = random.randint(0, 8)#создаем случайные индексы от 0 до 8
    j = random.randint(0, 8)
    chess_board[i][j] = piece#добавляем на случайное место на доске наши фигуры
```

```python
chess_board#посмотрим на полученную доску
```

А теперь нарисуем наше творение!

```python
from PIL import Image, ImageDraw, ImageFont
#страшная функция для отрисовки шахматной доски, пока что воспринимаем ее как черный ящик пандоры 
def draw_board(num, size, chess_board):
    new_color = (255, 255, 255) #белый цвет
    new_image = Image.new("RGB", (num * size, num * size), new_color)#заливаем все поле белым цветом
    x = size * num#создаем начало квадрата
    y = x#у квадрата стороны равны. поэтому координата по y=x
    draw = ImageDraw.Draw(new_image)#инициализируем рисовку на нашей картинке
    for i in range(0, x, size):#пробегаемся по массиву с шагом size(ширина квадрата)
        if i % (size * 2) == 0:#если номер квадарата четный заливаем его черным цветом
            for j in range(0, y, size):
                if j % (size * 2) == 0:
                    draw.rectangle(
                        [i, j, i + size - 1, j + size - 1], fill='black')#рисуем прямоугольник черного цвета
        else:
            for j in range(size, y, size):#если номер квадрата белый заливаем его белым цветом
                if j % (size * 2) != 0:
                    draw.rectangle(
                        [i, j, i + size - 1, j + size - 1], fill='black')#рисуем прямоугольник белого цвета
    for i in range(len(chess_board)):#пишем текст
        for j in range(len(chess_board)):
            draw.text((i*size+30, j*size+30), str(chess_board[i][j]), (0,255,0))#в соответствии с нашей расстановкой фигур пишем текст
    new_image.show()#отображаем картинку 
draw_board(8, 100, chess_board)
```

### Генераторы списков

Генератор списка позволяет создать список и заполнить его данными, используя всего одну строку кода. Выглядит это приблизительно так:

```python
my_list = []
for i in range(10):
    my_list.append(i)
my_list
```

```python
#альтернатива
my_list = [i for i in range(10)]
my_list
```

Генераторы списков - это гибрид цикла for и списка.

```python
squares = [x ** 2 for x in my_list] 

# равно по смыслу
for x in range(10):
    squares.append(x ** 2)

# но при этом гораздо быстрее и поэтому предпочтительно использовать именно генераторы списков для такого сценария
```

Генераторы могут содержать условие

```python
my_list = [x for x in range(10, 100) if x%3 == 0 and x%5 == 0]
my_list
```

## <h2><center>Словарь</center></h2>

Список хорош, когда он один. Но зачастую у нас возникает необходимость хранить несколько связанных списков. 
Например, номера телефонов и имена покупателей. Или название продуктов и их стоимость. В python есть особый тип данных для хранения такой информации в удобной форме - словарь

Словарь в Python - неупорядоченная ассоциативная изменяемая коллекция произвольных типов 

Словари иногда встречаются в других языках как "ассоциативные записи" или "ассоциативные массивы". В отличие от последовательностей, которые индексируются диапазоном чисел, словари индексируются по ключам, которые могут быть любым **неизменяемым** типом; строки и числа всегда могут быть ключами.

```python
fruit_dict = {'Банан': '16 рублей', 'Яблоко': '28 рублей', 'Персик': '37 рублей', 
             'Манго': '100 рублей', 'Апельсин': '30 рублей'}

#Обращаемся к элементу с помощью ключа
fruit_dict['Банан']
```

Получить ключи и значения словарей можно с помощью методов: keys(), values() и items()

```python
prices = list(fruit_dict.values())
prices
```

```python
list(fruit_dict.keys())
```

```python
list(fruit_dict.items())
```

```python
if 'Банан' not in fruit_dict : 
    print('Нет в словаре') 
else: 
    print(fruit_dict['Банан']) 
```

```python
print(fruit_dict.get('Апельсин', 0)) # надежный метод получения значения
# print(test_dict['Апельсин'])# ненадежный метод
```

### Циклы и словари

```python
for employee in employee_base:
    print(employee)
```

```python
for key, value in employee_base.items():
    print(key, value)
```
